# Read Data and Preprocess

<i><b> First read the file into program. And prepare each column as a data of list. 

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
f = open("stroopdata.csv")
data = [[float(line.split(',')[0]), float(line.split(',')[1])] for line in f.readlines()[1:]]
congruent = [x[0] for x in data]
incongruent = [x[1] for x in data]
print "The first column have "+str(len(congruent))+" numbers."
print "The second column have "+str(len(incongruent))+" numbers."
```

# Dependent and Independent Variables

<i><b>As we can see, the data contains two columns of data which are correspending to two tested cases in this experiment. There are 24 numbers in each column which are corresponding to 24 subjects in the research. These 24 subjects are tested in two cases. And these two cases as a binary variable is independent variable.

Case 1. Subjects are shown words of color to which are painted corresponding. For example, <font color='red'>red</font>, <font color='blue'>blue</font> and <font color='black'>black</font>.  

Case 2, Subjects are shown words of color to which are painted corresponding not. For example, <font color='red'>blue</font>, <font color='blue'>red</font> and <font color='green'>black</font>.

We will refer the paint of the word as inks and the word itself as word. We refer case 1 as congruent case and case 2 as incongruent case.

The length of time is measured for each subject naming the inks of a list of words which both in congruent and incongruent cases. The length of time as a numeric variable is dependent variable.

#  Hypothesis and Satistical Test

<i><b>The main question associated with this dataset is if there exists a difference for people making reaction with  congruent and incongruent situations. The people's reaction is measured by length of time in which subjects identify the inks of a list of words. We expected length of time to be different for people recognizing inks of congruent and incongruent list of words. This is general idea of my hypothesis test. Here is the detail.

I will use Tc for length of time it takes to name the inks in congruent words list and Ti for length of time it takes to name the inks in incongruent words list.

Null Hypothesis: Tc = Ti. The length of times are the same. The difference is due to random fluctuation and the length of time is not dependent on type of the words list(congruent and incongruent).

Alternative Hypothesis: Tc < Ti. The time naming the inks in incongruent words list is longer than that in congruent words. The time for people identify inks of words is dependent on type of the words list(congruent and incongruent). I eliminate the hypothesis in which Tc > Ti intuitively. Because prior knowledge about recognizing inks of words indicate that this is highly unlikely. 

I choose T-test of difference of times for this hypothesis. Because the sample is paired. That is every subject is exposed to two conditions. And the number of sample is relatively small. Dependent t-test is appropriate for this.

# Exploratory Data Analysis

<i><b>First, the means, variances and standard deviations are calculated for congruent and incongruent cases. The average length of time for incongruent list is longer than that for congruent list (22.16 vs 14.05).

```python
print "The average, variance and standard deviation of length of time for congruent list is \n"+str(np.mean(congruent))+", " + str(np.var(congruent)) + " and " + str(np.std(congruent)) + "."
print "The average, variance and standard deviation of length of time for incongruent list is \n"+str(np.mean(incongruent))+", " + str(np.var(incongruent)) + " and " + str(np.std(incongruent)) + "." 
```

<i><b>Mean, variance the and stand deviation of difference of two times is also calculated. The average difference is positive which is 7.96479166667.

```python
difference = [x[1] - x[0] for x in data]
```

```python
print "The average, variance and standard deviation of difference of time is \n"+str(np.mean(difference))+", " + str(np.var(difference, ddof  =1)) + " and " + str(np.std(difference, ddof  =1)) + "."
```

<i><b>The histogram and box-plot of length of time for congruent and incongruent are shown below. The centre of incongruent list is clearly highly deviated from that of congruent list.

```python
%pylab inline
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm
pylab.rcParams['figure.figsize'] = 12, 8
fs = 20
(mu_c, sigma_c) = norm.fit(congruent)
(mu_i, sigma_i) = norm.fit(incongruent)
n, bins, patches = plt.hist(congruent, 8, normed=1, facecolor='green', alpha=0.5)
y = mlab.normpdf( bins, mu_c, sigma_c)
l = plt.plot(bins, y, 'g--', linewidth=3)
n, bins, patches = plt.hist(incongruent, 8, normed=1, facecolor='red', alpha=0.5)
y = mlab.normpdf( bins, mu_i, sigma_i)
l = plt.plot(bins, y, 'r--', linewidth=3)
plt.xlabel('Times',size=fs)
plt.ylabel('Probability',size=fs)
plt.legend(('congruent', 'incongruent'),prop={'size':fs})
plt.grid(True)
plt.show()

plt.boxplot([congruent, incongruent], labels = ['congruent', 'incongruent'])
plt.xticks([1, 2], ['congruent', 'incongruent'], size=fs )
plt.ylabel('Times',size=fs)
plt.show()
```

# Results of Statistical Test

<i><b>This is a t-test of difference, I set alpha level at 0.01. Here is calculation result of this test:

Alpha level: 0.01

T-critical Value: 3.485 

Mean: 7.965

Number of Sample: 24

Degree of Freedom: 23

Standard Deviation: 4.864

Margin of Error: 0.993

T-statistics: 8.021

Since T-statistics is far passed T-critical value(8.021 vs 3.485). We will reject null hypothesis. We think that time for people identifying incongruent list is longer than for congruent list is a statistically significant result. 

This result showed that people may have slower reaction for incongruent list of words.
